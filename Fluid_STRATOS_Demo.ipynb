## 1. ğŸŒ€ Initializing the Cognitive Field

Let's create a Fluid STRATOS instance with a 2D cognitive wave function.

```python
from fluid_stratos import FluidSTRATOS
import numpy as np

# Create the cognitive field (64x64 grid)
print("ğŸŒ€ Initializing cognitive field...")
stratos = FluidSTRATOS(grid_size=(64, 64))

print(f"\nâœ… Cognitive field created!")
print(f"   Grid size: {stratos.grid_size}")
print(f"   Number of modes: 16")
print(f"   Initial brain energy: {stratos.brain_energy:.4f}")
```

## 2. ğŸ§  Exploring 16 Cognitive Modes

The 16 modes represent different aspects of cognition, inspired by the Hope Genome.

```python
# The 16 cognitive modes
mode_names = [
    "Brain", "Heart", "Logic", "Emotion",
    "Memory", "Imagination", "Intuition", "Reasoning",
    "Perception", "Action", "Language", "Creativity",
    "Planning", "Reflection", "Social", "Meta-cognition"
]

print("ğŸ§  The 16 Cognitive Modes:\n")
for i, name in enumerate(mode_names):
    print(f"  Mode {i:2d}: {name}")

# Excite specific modes
print("\nğŸ’« Exciting modes...")
stratos.excite_mode(0, strength=2.0)  # Brain
stratos.excite_mode(6, strength=1.5)  # Intuition
stratos.excite_mode(11, strength=1.0) # Creativity

print("   Brain mode excited (strength: 2.0)")
print("   Intuition mode excited (strength: 1.5)")
print("   Creativity mode excited (strength: 1.0)")
```

## 3. ğŸŒŠ Evolving the Cognitive Field

Let's evolve the system and watch the wave function flow according to the Gross-Pitaevskii Equation.

```python
print("ğŸŒŠ Evolving cognitive field...")

# Evolve for 100 timesteps
for step in range(100):
    stratos.evolve(steps=1)
    if step % 20 == 0:
        energy = stratos.brain_energy
        print(f"   Step {step:3d}: Brain energy = {energy:.4f}")

print("\nâœ… Evolution complete!")
print(f"   Final brain energy: {stratos.brain_energy:.4f}")
```

## 4. ğŸ¼ Hope Genome Vote

The Hope Genome Vote shows which cognitive modes are currently dominant.

```python
# Get the Hope Genome Vote
vote = stratos.hope_genome_vote()

print("ğŸ¼ Hope Genome Vote Results:\n")
print(f"  Total energy: {vote['total_energy']:.4f}")
print(f"  Coherence: {vote['coherence']:.4f}")
print(f"  Is resonating: {vote['is_resonating']}")
print(f"\n  Top 5 dominant modes:")

for mode_idx in vote['dominant_modes'][:5]:
    energy = vote['mode_energies'][mode_idx]
    print(f"    {mode_names[mode_idx]:15s} (Mode {mode_idx:2d}): {energy:.4f}")
```

## 5. ğŸ’¾ Creating Emotional Memories (EmotiMem)

EmotiMem stores memories as wave packets with emotional valence.

```python
print("ğŸ’¾ Creating emotional memories...\n")

# Store some memories
memories = [
    {"position": (32, 32), "valence": 0.8, "tag": "happy_moment"},
    {"position": (20, 20), "valence": -0.3, "tag": "challenge"},
    {"position": (45, 45), "valence": 0.5, "tag": "discovery"}
]

for mem in memories:
    stratos.store_memory(
        position=mem["position"],
        valence=mem["valence"],
        tag=mem["tag"]
    )
    print(f"  Stored: {mem['tag']:15s} at {mem['position']} (valence: {mem['valence']:+.2f})")

print(f"\nâœ… {len(memories)} memories stored!")
print(f"   Total memories in system: {len(stratos.memory_traces)}")
```

## 6. ğŸ¨ Visualizing the Cognitive Field

Let's visualize the cognitive wave function and see the flow of thought.

```python
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm

# Get the wave function magnitude
psi = stratos.psi
magnitude = np.abs(psi)**2

# Create visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Wave function magnitude
im1 = axes[0].imshow(magnitude, cmap='viridis', origin='lower')
axes[0].set_title('ğŸŒŠ Wave Function Magnitude')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
plt.colorbar(im1, ax=axes[0])

# Wave function phase
phase = np.angle(psi)
im2 = axes[1].imshow(phase, cmap='twilight', origin='lower')
axes[1].set_title('ğŸ¨ Wave Function Phase')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
plt.colorbar(im2, ax=axes[1])

# Potential landscape
V = stratos.V
im3 = axes[2].imshow(V, cmap='RdYlBu_r', origin='lower')
axes[2].set_title('â›°ï¸ Potential Landscape')
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')
plt.colorbar(im3, ax=axes[2])

# Mark memory locations
for mem in memories:
    y, x = mem["position"]
    color = 'lime' if mem["valence"] > 0 else 'red'
    axes[0].scatter(x, y, c=color, s=100, marker='*', edgecolors='white', linewidths=1)

plt.tight_layout()
plt.show()

print("\nğŸ“Š Visualization complete!")
```

## 7. ğŸŒ± Cognitive Gardener

Let's add a Cognitive Gardener to maintain homeostatic balance.

```python
from cognitive_gardener import CognitiveGardener

# Initialize gardener
print("ğŸŒ± Initializing Cognitive Gardener...")
gardener = CognitiveGardener(stratos, target_brain_energy=0.25)

print("\nğŸŒ± Running homeostatic control...")
energies = []

for t in range(50):
    stratos.evolve(steps=1)
    gardener.step()
    energies.append(stratos.brain_energy)
    if t % 10 == 0:
        print(f"   Step {t:2d}: Energy = {stratos.brain_energy:.4f}")

# Plot energy over time
plt.figure(figsize=(10, 4))
plt.plot(energies, label='Brain Energy', linewidth=2)
plt.axhline(y=0.25, color='r', linestyle='--', label='Target Energy', linewidth=2)
plt.xlabel('Time Step')
plt.ylabel('Brain Energy')
plt.title('ğŸŒ± Cognitive Gardener - Homeostatic Regulation')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print("\nâœ… Homeostatic control complete!")
```
